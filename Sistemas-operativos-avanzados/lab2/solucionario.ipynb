### Solución 3

MLFQ se aproximará a un RR cuando hay una sola cola y el quantum sea menor que la longitud del trabajo más grande dividida entre la cantidad total de trabajos. Así, el planificador cambiará entre trabajos de la misma forma que en el round-robin.

### Solución 5

Necesitamos que nuestro trabajo largo se ejecute al menos 5 veces de cada 100. Sin embargo, esto no es posible de forma exacta dado nuestro cuanto, 10ms. Tendríamos 

| B(ms)|ti100|
|------|-----|
|  10  |  100|
|  20  |  40 |
|  30  |  30 |
|  40  |  20 |
|  50  |  20 |
|  60  |  10 |

Así, lo que está más cerca de ese límite será 40ms o 50ms. Si tomamos una muestra más grande, convidaremos en que el que cumple es solamente 50ms puesto que hay más densidad de múltiplos de 4 a medida entre 0 y 100 que entre 0 y 10
